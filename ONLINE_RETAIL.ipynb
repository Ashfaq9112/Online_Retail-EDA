# Creating New column called 'Recency' to calculate recency rate

```python
df['Recency'] = (current_date - df['InvoiceDate']).dt.days
```

```python
least_recency_values = df.groupby('Country')['Recency'].min()
result_df = pd.DataFrame({'Country': least_recency_values.index, 'Least_Recency': least_recency_values.values})
result_df.sort_values(by='Least_Recency',ascending=True).head(50)
```

```python
most_recency_values = df.groupby('Country')['Recency'].max()
result_df1 = pd.DataFrame({'Country':most_recency_values.index, 'Most_Recency': most_recency_values.values})
result_df1.sort_values(by='Most_Recency',ascending=False).head(50)
```

# The table below shows least and most recent days of each countries

1. Almost every country has high nuumber of most recency data
2. In Brazil,Lebanon,bahrain, Lebanon, Saudi Arabia has highes least recency rate
3. RSA, European Community and Czech Republic has less recency because the sales happens only on those months
4. Countries like EIRE, UK, France, Germany and Norway has 0 recency rate this tells us that the shops are running regularly
5. Other countries with one digit rececny rate are running regularly

```python
result_df= pd.merge(result_df,result_df1, on='Country')
result_df
```

# Sales in each country by month

```python
countries = df['Country'].unique()
for country in countries:
    country_data = df[df['Country'] == country]

    monthly_sales = country_data.groupby('month')['Amount'].sum()

    plt.figure(figsize=(10, 5))
    plt.bar(monthly_sales.index, monthly_sales.values)
    plt.xlabel('Month')
    plt.ylabel('Sales')
    plt.title(f'Sales in Each Month for {country}')
    plt.show()
```
